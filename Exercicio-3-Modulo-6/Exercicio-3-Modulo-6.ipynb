```python
# Importando bibliotecas.
from IPython.display import display # biblioteca para exibir um dataframe de forma mais apresentável quando chamando de uma função. (Item 4 deste notebook)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# Configurando a visualização para poder ver todas as colunas e linhas do dataframe.
pd.set_option('display.max_columns', None)
pd.set_option("display.max_rows", None)

# Importando os dados.
df_datasus = pd.read_csv('sinasc_RO_2019.csv')

# Visualizando primeiras linhas.
df_datasus.head()
```

```python
# Verificando o tamanho e algumas estatíticas para entender um pouco melhor a base de dados.
print("A base de dados tem o seguinte número de (linhas, colunas): ", df_datasus.shape)
df_datasus.describe()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Criando a coluna de categorização para a variável APGAR5.
df_datasus.loc[(df_datasus['APGAR5'] >= 0) & (df_datasus['APGAR5'] <= 3), 'APGAR5_CAT'] = 'asfixia grave'
df_datasus.loc[(df_datasus['APGAR5'] >= 4) & (df_datasus['APGAR5'] <= 5), 'APGAR5_CAT'] = 'asfixia moderada'
df_datasus.loc[(df_datasus['APGAR5'] >= 6) & (df_datasus['APGAR5'] <= 7), 'APGAR5_CAT'] = 'asfixia leve'
df_datasus.loc[(df_datasus['APGAR5'] >= 8) & (df_datasus['APGAR5'] <= 10), 'APGAR5_CAT'] = 'normal'

# Exibindo as colunas APGAR5 e APGAR5_CAT para validar a criação da nova coluna.
df_datasus[['APGAR5', 'APGAR5_CAT']].head()
```

```python
# Função para criar gráfico de distribuição.
def criar_grafico_distribuicao(dataframe: pd.DataFrame, coluna_eixo_x: str, coluna_eixo_y: str, coluna_divisor_categorico: str, titulo_grafico: str, rotulo_eixo_x: str, rotulo_eixo_y: str, titulo_legenda: str):
    # Criando o gráfico
    plt.figure(figsize=(12, 6))
    sns.boxplot(data=dataframe, x=coluna_eixo_x, y=coluna_eixo_y, hue=coluna_divisor_categorico)

    # Adicionando títulos e rótulos
    plt.title(titulo_grafico, fontsize=16)
    plt.xlabel(rotulo_eixo_x, fontsize=12)
    plt.ylabel(rotulo_eixo_y, fontsize=12)

    # Exibindo o gráfico
    plt.legend(title=titulo_legenda, loc="upper left", bbox_to_anchor=(1, 1))
    plt.tight_layout()
    plt.show()
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# Gerando o gráfico usando a coluna APGAR5.
criar_grafico_distribuicao(df_datasus, "GESTACAO", "APGAR5", "SEXO", "Distribuição do APGAR5 por Sexo e Tempo de Gestação", "Tempo de Gestação", "APGAR5", "Sexo")

# Gerando o gráfico usando a coluna APGAR5_CAT (Coluna de categorização do APGAR5 criada no item 2).
criar_grafico_distribuicao(df_datasus, "GESTACAO", "APGAR5_CAT", "SEXO", "Distribuição do APGAR5 (categorizado) por Sexo e Tempo de Gestação", "Tempo de Gestação", "APGAR5_CAT", "Sexo")
```

```python
# Função para criar uma tabela cruzada.
def criar_tabela_cruzada(serie_eixo_x: pd.core.series.Series, serie_eixo_y: pd.core.series.Series) -> pd.DataFrame:
    df_tabela_cruzada = pd.crosstab(serie_eixo_x, serie_eixo_y)
    return df_tabela_cruzada
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Criando a tabela cruzada entre APGAR5 e GESTACAO.
print("Tabela cruzada entre APGAR5 e GESTACAO:")
df_apgar5_gestacao = criar_tabela_cruzada(df_datasus['GESTACAO'], df_datasus['APGAR5'])
display(df_apgar5_gestacao)

# Criando a tabela cruzada entre APGAR5_CAT e GESTACAO.
print("Tabela cruzada entre APGAR5_CAT e GESTACAO:")
df_apgar5_cat_gestacao = criar_tabela_cruzada(df_datasus['GESTACAO'], df_datasus['APGAR5_CAT'])
display(df_apgar5_cat_gestacao)
```

```python
# Função para criar uma tabela cruzada percentual.
def criar_tabela_cruzada_percentual(dataframe: pd.DataFrame) -> pd.DataFrame:
    df_percentual = dataframe.div(dataframe.sum(axis = 1), axis = 0)
    return df_percentual
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Criando a tabela cruzada (percentual) entre APGAR5 e GESTACAO.
print("Tabela cruzada (percentual) entre APGAR5 e GESTACAO:")
df_apgar5_gestacao_percentual = criar_tabela_cruzada_percentual(df_apgar5_gestacao)
display(df_apgar5_gestacao_percentual)
print("Validando os percentuais. O valor esperado é um (1) para cada linha. Isso indica que o somatório das colunas de cada linha é 100%.")
display(df_apgar5_gestacao_percentual.sum(axis = 1))

# Criando a tabela cruzada (percentual) entre APGAR5_CAT e GESTACAO.
print("Tabela cruzada (percentual) entre APGAR5_CAT e GESTACAO:")
df_apgar5_cat_gestacao_percentual = criar_tabela_cruzada_percentual(df_apgar5_cat_gestacao)
display(df_apgar5_cat_gestacao_percentual)

# Validando a tabela cruzada (percentual) entre APGAR5_CAT e GESTACAO.
print("Validando os percentuais. O valor esperado é um (1) para cada linha. Isso indica que o somatório das colunas de cada linha é 100%.")
display(df_apgar5_cat_gestacao_percentual.sum(axis = 1))
```

```python
# Função para criar um gráfico de barras.
def criar_grafico_barras(dataframe: pd.DataFrame, texto_legenda: str) -> None:
    eixos = dataframe.plot.bar()
    eixos.legend(title=texto_legenda, loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Abaixo são criados 2 gráficos. Particularmente, eu prefiro o segundo gráfico, pois ele é mais fácil de interpretar dada a categorização do APGAR5.

# Criando o gráfico de barras para ver a relação entre APGAR5 e o tempo de gestação
criar_grafico_barras(df_apgar5_gestacao_percentual, "APGAR5")

# Criando o gráfico de barras para ver a relação entre APGAR5 (categorizado) e o tempo de gestação
criar_grafico_barras(df_apgar5_cat_gestacao_percentual, "APGAR5 (categorizado)")
```

```python
# Resolvi criar uma coluna nova, PPREMATURO, e a partir dela comparar com a coluna APGAR5_CAT.
# Através dela percebo que o número de casos de crianças prematuras é bem menor quando o valor do APGAR5 categorizado é considerado normal.

# Criando a coluna PREMATURO, com base na coluna GESTACAO, para facilitar avaliar se um bebê é prematuro ou não.
df_datasus.loc[(df_datasus['GESTACAO'] == '42 semanas e mais'), 'PREMATURO'] = 'Não'
df_datasus.loc[(df_datasus['GESTACAO'] == '37 a 41 semanas'), 'PREMATURO'] = 'Não'
df_datasus.loc[(df_datasus['GESTACAO'] == '32 a 36 semanas'), 'PREMATURO'] = 'Sim'
df_datasus.loc[(df_datasus['GESTACAO'] == '28 a 31 semanas'), 'PREMATURO'] = 'Sim'
df_datasus.loc[(df_datasus['GESTACAO'] == '22 a 27 semanas'), 'PREMATURO'] = 'Sim'
df_datasus.loc[(df_datasus['GESTACAO'] == 'Menos de 22 semanas'), 'PREMATURO'] = 'Sim'

# Criando a tabela cruzada entre APGAR5_CAT e a nova coluna PREMATURO.
df_apgar5_cat_prematuro = criar_tabela_cruzada(df_datasus['APGAR5_CAT'], df_datasus['PREMATURO'])

# Criando a tabela cruzada (percentual) entre APGAR5_CAT e a nova coluna PREMATURO.
df_apgar5_cat_prematuro_percentual = criar_tabela_cruzada_percentual(df_apgar5_cat_prematuro)

# Criando o gráfico de barras para ver a relação entre APGAR5 (categorizado) e o tempo de gestação
criar_grafico_barras(df_apgar5_cat_prematuro_percentual, "Prematuro")
```

```python
# Criando um gráfico do tipo boxplot para visualizar a distribuição do PESO por APGAR5 (categorizado).
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_datasus, x="APGAR5_CAT", y="PESO", palette="Set2", hue="APGAR5_CAT")

# Configurando rótulos
plt.title("Distribuição do PESO por APGAR5 (Categorizado)", fontsize=16)
plt.xlabel("Categoria de APGAR5", fontsize=12)
plt.ylabel("Peso (gramas)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
#Criando um gráfico do tipo violinplot para visualizar a distribuição do PESO por APGAR5 (categorizado).
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_datasus, x="APGAR5_CAT", y="PESO", palette="coolwarm", inner="quartile", hue="APGAR5_CAT")

# Configurando rótulos
plt.title("Distribuição do PESO por APGAR5 (Categorizado)", fontsize=16)
plt.xlabel("Categoria de APGAR5", fontsize=12)
plt.ylabel("Peso (gramas)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Criando um gráfico do tipo Stripplot para visualizar a distribuição do PESO por APGAR5 (categorizado).
plt.figure(figsize=(10, 6))
eixos = sns.stripplot(data=df_datasus, x="APGAR5_CAT", y="PESO", palette="Set1", jitter=True, hue="APGAR5_CAT")
eixos.legend(title="APGAR5_CAT", loc="upper left", bbox_to_anchor=(1, 1))

# Configurando rótulos
plt.title("Distribuição do PESO por APGAR5 Categorizado (com pontos individuais)", fontsize=16)
plt.xlabel("Categoria de APGAR5", fontsize=12)
plt.ylabel("Peso (gramas)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
