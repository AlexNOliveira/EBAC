#### Marcação de bom e mau
O objetivo da modelagem é classificar o risco de inadimplência, ou como se diz no meio, o risco de *default*. Podemos fazer longas discussões sobre o conceito de *default* com base em estudos e exigências regulatórias, para efeitos deste estudo, um cliente em *default* é aquele que está em 60 dias de atraso ou mais. Então classificaremos os clientes como 'bons' e 'maus' assim:
- **Maus** pagadores: são aqueles que entraram em 'default' (atraso 60 dias ou mais) nos 24 meses seguintes à aquisição do cartão de crédito. 
- **Bons** pagadores: são considerados todos os demais.
- **Excluídos**: Clientes que não adquiriram um cartão de crédito (seja por recusa, seja por desistência) não possuem informações de pagamento, portanto não se pode identificar se são bons ou maus. Há uma longa discussão e literatura sobre *inferência de rejeitados* que está fora do escopo deste exercício.

#### Construindo a variável resposta
A base de pagamentos está em um formato de 'base larga'. Essa base possui informações de pagamentos do cliente mês a mês a partir do mês de aquisição do crédito (mês 0) até o vigésimo quarto mês após a aquisição do crédito (mês 24). Utilizaremos essa base para determinar se um proponente é considerado 'bom pagador' ou caso apresente atraso representativo, será considerado 'mau pagador'.

#### Base larga vs base longa
A base ser larga significa que há uma linha para cada cliente, e que as informações estarão nas colunas, em contraste com a 'base longa', em que haveria uma linha para cada combinação cliente/mês, uma coluna indicando o cliente, outra indicando o mês, e apenas uma coluna com a informação do atraso.

````python
import pandas as pd
````

````python
propostas = pd.read_csv('application_record.csv')
pg = pd.read_csv('pagamentos_largo.csv')
````

````python
print('Visualizando 10 linhas das propostas...')
propostas.head(10)
````

````python
print('Visualizando 10 linhas dos pagamentos...')
pg.head(10)
````

````python
# 1) Seu código aqui

default_columns = [f'mes_{i}' for i in range(25)] 
print('Iniciando a criação e a atribuição de valor das colunas default para cade mês:\n')
for col in default_columns:
   print(f" Criando e populando a coluna 'default_{col}'.")
   pg[f'default_{col}'] = pg[col].isin([2,3,4,5]) 

# Visualizar as primeiras linhas do DataFrame com as novas colunas de default
print('\nApresentando as primeiras 10 linhas do dataframe.')
pg.head(10)
````

````python
# 2) Seu código aqui

# Meu entendimento foi de que o que foi descrito no vídeo e o que é pedido aqui estão diferentes e por isso escolhi seguir o que estava descrito no vídeo. (Criei apenas uma coluna ao invés de substituir os valores nas colunas de cada mês)
# Escolhi a abordagem de quantidade de ocorrências ao invés de True ou False pensando em uma possibilidade futura de analizar o nível de risco do cliente e não somente se ele representa um risco ou não.
print("Criando uma coluna 'default' geral e atribuindo a ela a quantidade de vezes (meses) que a pessoa ficou 60 dias ou mais sem pagar.")
print("O valor zero na coluna 'default' significa que sempre pagou em dia ou com menos de 60 dias de atraso.")
pg['default'] = pg.isin(['2', '3', '4', '5']).sum(axis=1)
pg_filtrado = pg.copy()
print('Limpando o dataframe de pagamentos, removendo as linhas que não tenham todos os 24 meses preenchidos.\nEssa limpeza é necessária pois sem todos os meses preenchidos não é possível determinar com exatidão se o cliente atrasou mais do que 60 dias.')
pg_filtrado_sem_nulos = pg_filtrado.dropna().copy()
pg_filtrado_sem_nulos
````

````python
# 3) Seu código aqui

# Filtrar apenas clientes que estão na base de pagamentos e que tenham todos os dados preenchidos.
print('Criando dataframe com proponentes expostos a risco.\n')
clientes_com_cartao = propostas.merge(pg, on='ID').dropna()

# Verificar as primeiras linhas
print('Apresentando as primeiras 10 linhas do dataframe.')
clientes_com_cartao.head(10)
````

````python
# 4) Seu código aqui

# Fazer a junção e manter apenas os clientes que adquiriram cartão
print('Criando dataframe para consolidar somente os clientes com cartão.\n')
clientes_com_informacoes = propostas.merge(clientes_com_cartao[['ID', 'default']], on='ID')

# Verificar as primeiras linhas para confirmar
print('Apresentando as primeiras 10 linhas do dataframe.')
clientes_com_informacoes.head(10)
````

````python
# 5) Seu código aqui

quantidade_clientes_tomadores_maus, _ = clientes_com_informacoes[ clientes_com_informacoes['default'] >= 1 ].shape
quantidade_clientes_tomadores_bons, _ = clientes_com_informacoes[ clientes_com_informacoes['default'] == 0 ].shape

print(f"Temos um total de {quantidade_clientes_tomadores_maus} clientes que adquiriram cartão de crédito e representam um risco.")
print(f"Temos um total de {quantidade_clientes_tomadores_bons} clientes que adquiriram cartão de crédito e que não representam um risco.")
````
