```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
df_datasus = pd.read_csv('sinasc_RO_2019.csv')
df_datasus.head()
```

##### Criando função para montar um gráfico de barras

```python
def montar_grafico_de_barras(dicionario: dict, titulo_principal: str, titulo_eixo_x: str, titulo_eixo_y: str, graus_rotacao_titulo_eixo_x: int = 0) -> None:
    
    descricoes = list(dicionario.keys())
    valores = list(dicionario.values())
 
    fig = plt.figure(figsize = (10, 5))

    plt.bar(descricoes, valores)

    plt.title(titulo_principal)
    plt.xlabel(titulo_eixo_x)
    plt.ylabel(titulo_eixo_y)
    plt.xticks(descricoes, rotation=graus_rotacao_titulo_eixo_x)
    plt.show()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Criando colunas de descrição para ESTCIVMAE e CONSULTAS. Serão usadas posteriormente.

# Incluindo coluna descritiva para o estado civil da mãe

df_datasus.loc[df_datasus['ESTCIVMAE'] == 1, 'ESTCIVMAE_DESC'] = 'Solteira'
df_datasus.loc[df_datasus['ESTCIVMAE'] == 2, 'ESTCIVMAE_DESC'] = 'Casada'
df_datasus.loc[df_datasus['ESTCIVMAE'] == 3, 'ESTCIVMAE_DESC'] = 'Viúva'
df_datasus.loc[df_datasus['ESTCIVMAE'] == 4, 'ESTCIVMAE_DESC'] = 'Separada judicialmente/Divorciada'
df_datasus.loc[df_datasus['ESTCIVMAE'] == 5, 'ESTCIVMAE_DESC'] = 'União consensual (versões anteriores)'
df_datasus.loc[df_datasus['ESTCIVMAE'] == 9, 'ESTCIVMAE_DESC'] = 'Ignorado'

# Incluindo coluna descritiva para o número de consultas realizadas

df_datasus.loc[df_datasus['CONSULTAS'] == 1, 'CONSULTAS_DESC'] = 'Nenhuma'
df_datasus.loc[df_datasus['CONSULTAS'] == 2, 'CONSULTAS_DESC'] = 'de 1 a 3'
df_datasus.loc[df_datasus['CONSULTAS'] == 3, 'CONSULTAS_DESC'] = 'de 4 a 6'
df_datasus.loc[df_datasus['CONSULTAS'] == 4, 'CONSULTAS_DESC'] = '7 e mais'
df_datasus.loc[df_datasus['CONSULTAS'] == 9, 'CONSULTAS_DESC'] = 'Ignorado'

# Exibindo todas as colunas para validar a criação das novas colunas.abs

df_datasus.head()
```

```python
# Criando um dicionário para cada uma das variáveis solicitadas.

dicionario_apgar1 = df_datasus['APGAR1'].value_counts().to_dict()
dicionario_apgar5 = df_datasus['APGAR5'].value_counts().to_dict()
dicionario_sexo = df_datasus['SEXO'].value_counts().to_dict()
dicionario_gestacao = df_datasus['GESTACAO'].value_counts().to_dict()

# Usando a coluna descritiva para melhor interpretação do gráfico.
dicionario_estado_civil_mae = df_datasus['ESTCIVMAE_DESC'].value_counts().to_dict()

# Executando a função de montagem do gráfico para cada variável. 

montar_grafico_de_barras(dicionario=dicionario_apgar1, 
                         titulo_principal='Escala Apgar - Primeiro Minuto', 
                         titulo_eixo_x='Pontuação', 
                         titulo_eixo_y='Quantidade de Nascidos')

montar_grafico_de_barras(dicionario=dicionario_apgar5, 
                         titulo_principal='Escala Apgar - Quinto Minuto', 
                         titulo_eixo_x='Pontuação', 
                         titulo_eixo_y='Quantidade de Nascidos')

montar_grafico_de_barras(dicionario=dicionario_sexo, 
                         titulo_principal='Sexo do Nascido', 
                         titulo_eixo_x='Gênero', 
                         titulo_eixo_y='Quantidade de Nascidos')

montar_grafico_de_barras(dicionario=dicionario_gestacao, 
                         titulo_principal='Gestação', 
                         titulo_eixo_x='Tempo de Gestação', 
                         titulo_eixo_y='Quantidade de Nascidos', 
                         graus_rotacao_titulo_eixo_x=30)

montar_grafico_de_barras(dicionario=dicionario_estado_civil_mae, 
                         titulo_principal='Estado Civil da Mãe', 
                         titulo_eixo_x='Estado Civil', 
                         titulo_eixo_y='Quantidade de Nascidos', 
                         graus_rotacao_titulo_eixo_x=75)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
dicionario_apgar5_medio_por_consulta = df_datasus.groupby('CONSULTAS')['APGAR5'].mean().to_dict()

montar_grafico_de_barras(dicionario=dicionario_apgar5_medio_por_consulta, 
                         titulo_principal='APGAR5 Médio por Consulta', 
                         titulo_eixo_x='Valor do APGAR5', 
                         titulo_eixo_y='Média de Consultas')
```

```python
def montar_grafico_de_frequencias(df, colunas, titulo_colunas, titulo, rotacao_x=0, limite_rotulo=0):

    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    fig.suptitle(titulo, fontsize=16)

    for i, coluna in enumerate(colunas):
        ax = axes[i // 2, i % 2]
        barras = df[coluna].value_counts().plot(kind='bar', ax=ax)
        ax.set_title(f'Frequência de {titulo_colunas[i]}')
        ax.set_xlabel(titulo_colunas[i])
        ax.set_ylabel('Contagem')
        ax.tick_params(axis='x', rotation=rotacao_x)

        for barra in barras.patches:
            height = barra.get_height()
            if height > limite_rotulo[i]:
                ax.annotate(
                    f'{int(height)}', 
                    xy=(barra.get_x() + barra.get_width() / 2, height),
                    xytext=(0, -10),
                    textcoords="offset points",
                    ha='center', va='center', 
                    fontsize=10, color='black'
                )
            else:
                ax.annotate(
                    f'{int(height)}', 
                    xy=(barra.get_x() + barra.get_width() / 2, height),
                    xytext=(0, 5),
                    textcoords="offset points",
                    ha='center', va='bottom', 
                    fontsize=10, color='black'
                )

    plt.tight_layout()
    plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
colunas = ['SEXO', 'GESTACAO', 'ESTCIVMAE_DESC', 'CONSULTAS_DESC']
titulo_colunas = ['Sexo', 'Gestação', 'Estado Civil', 'Consultas Realizadas']
limite_rotulo = [10000, 5000, 500, 1500]
montar_grafico_de_frequencias(df_datasus, colunas, titulo_colunas, 'Frequência das Variáveis', 80, limite_rotulo)
```
