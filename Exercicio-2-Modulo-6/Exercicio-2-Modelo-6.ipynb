```python
# Importando bibliotecas.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# Configurando a visualização para poder ver todas as colunas.
pd.set_option('display.max_columns', None)
# Importando os dados.
df_datasus = pd.read_csv('sinasc_RO_2019.csv')
# Visualizando primeiras linhas.
df_datasus.head()
```

```python
# Verificando o tamanho e algumas estatíticas para entender um pouco melhor a base de dados.
print("A base de dados tem o seguinte número de (linhas, colunas): ", df_datasus.shape)
df_datasus.describe()
```

```python
# Funções auxiliares para boxplot e histograma.

# Estou explicitamente usando o dataframe dentro das funções ao invés de passá-lo como variável.
# Foi um escolha consciente para simplificar dado que todos os gráficos serão em cima do mesmo dataframe.

def boxplot(coluna: str, legenda_eixo_y: str):
    bp_ax = sns.boxplot(y=coluna, data=df_datasus[[coluna]])
    bp_ax.set_ylabel(legenda_eixo_y)
    plt.plot()

def histograma(coluna: str, legenda_eixo_x: str, legenda_eixo_y: str, legendas: str, numero_de_quebras: int = 50):
    hist_ax = df_datasus[[coluna]].plot.hist(bins=numero_de_quebras)
    hist_ax.set_xlabel(legenda_eixo_x)
    hist_ax.set_ylabel(legenda_eixo_y)
    plt.legend(labels=[legendas])
    plt.plot()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
#Criando um boxplot para a idade da mãe.
boxplot('IDADEMAE', 'Idade da Mãe')
```

####  2) Faça um boxplot do peso do bebê

```python
#Criando um boxplot para o peso do bebê ao nascer.
boxplot('PESO', 'Peso em Gramas')
```

####  3) Faça um histograma da idade da mãe

```python
#Criando um histograma para a idade da mãe.
histograma('IDADEMAE', 'Idade da Mãe', 'Total de Mães', 'Idade', 20)
```

#### 4) Faça um histograma do peso do bebê

```python
#Criando um histograma para a idade da mãe.
histograma('PESO', 'Peso em Gramas', 'Total de Bebês', 'Peso', 40)
```
